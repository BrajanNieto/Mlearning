### 3. Desarrollo y Entrenamiento de Modelos:

Modelos de Regresión Lineal y Polinómica: Implementar modelos de regresión lineal simples y polinómica para establecer una línea base de desempeño.
Aplicación de Regularización:

Regresión Lasso (L1): Implementar regresión con regularización L1 para realizar selección de variables, identificando así las características más relevantes.

Regresión Ridge (L2): Implementar regresión con regularización L2 para controlar la magnitud de los coeficientes y evitar el sobreajuste.

Entrenamiento de Modelos: Ajustar los modelos utilizando el conjunto  de entrenamiento, optimizando hiperparámetros como el grado polinómico y el parámetro de regularización (alpha).

```python
# 1) Escalar X
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_val_scaled   = scaler_X.transform(X_val)

# 2) Ajustar regresión lineal simple
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# 3) Predecir y evaluar con MSE
y_pred = model.predict(X_val_scaled)
mse = mean_squared_error(y_val, y_pred)
print(f"LinearRegression (grado=1) → MSE: {mse:.4f}")
```

```python
# Grados a explorar
degrees = [2, 3, 4, 5]

# Escaladores (separados para X e Y)
scaler_X = StandardScaler()
scaler_y = StandardScaler()

# Escalamos Y fuera del bucle, para no re-ajustar el scaler cada vez
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).ravel()
y_val_scaled   = scaler_y.transform(y_val.values.reshape(-1, 1)).ravel()

best_deg = None
best_mse = np.inf

for deg in degrees:
    # 1) Generar polinomios de grado 'deg'
    poly = PolynomialFeatures(degree=deg, include_bias=False)
    X_train_poly = poly.fit_transform(X_train)
    X_val_poly   = poly.transform(X_val)

    # 2) Escalar el espacio polinómico
    X_train_scaled = scaler_X.fit_transform(X_train_poly)
    X_val_scaled   = scaler_X.transform(X_val_poly)

    # 3) Ajustar regresión lineal sobre X escalada y Y escalada
    model = LinearRegression()
    model.fit(X_train_scaled, y_train_scaled)

    # 4) Predecir, desescalar y calcular MSE en la escala original de Y
    y_pred_scaled = model.predict(X_val_scaled)
    y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()
    mse = mean_squared_error(y_val, y_pred)

    print(f"  Polinomial (grado={deg}) → MSE: {mse:.4f}")

    if mse < best_mse:
        best_mse, best_deg = mse, deg

print(f"> Mejor grado polinómico: {best_deg} con MSE {best_mse:.4f}")
```

```python
# 0) Define grados y alphas
degrees = [2, 3, 4, 5]
alphas = [1e-3, 1e-2, 1e-1, 1, 10]

# 1) Escala las X originales
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_val_scaled   = scaler_X.transform(X_val)

results = {}

for name, Model in [("Lasso", Lasso), ("Ridge", Ridge)]:
    best = {"degree": None, "alpha": None, "mse": np.inf}

    for deg in degrees:
        # 2) Genera polinomios sobre X escalada
        poly = PolynomialFeatures(degree=deg, include_bias=False)
        Xtr_poly = poly.fit_transform(X_train_scaled)
        Xvl_poly = poly.transform(X_val_scaled)

        # 3) Escala el espacio polinómico
        scaler_pf = StandardScaler()
        Xtr_poly_scaled = scaler_pf.fit_transform(Xtr_poly)
        Xvl_poly_scaled = scaler_pf.transform(Xvl_poly)

        # 4) Barra alphas
        for alpha in alphas:
            model = Model(alpha=alpha, max_iter=10_000)
            model.fit(Xtr_poly_scaled, y_train)
            y_pred = model.predict(Xvl_poly_scaled)
            mse = mean_squared_error(y_val, y_pred)

            if mse < best["mse"]:
                best.update({"degree": deg, "alpha": alpha, "mse": mse})

    results[name] = best

# 5) Imprime resultados
for name, r in results.items():
    print(f"{name}: mejor grado={r['degree']}, alpha={r['alpha']}, MSE={r['mse']:.4f}")
```

### 4. Evaluación y Comparación de Modelos:

Métricas de Desempeño: Calcular métricas como MSE (Mean Squared Error), RMSE (Root Mean Squared Error), R y R² tanto en los conjuntos de entrenamiento como de validación.

Comparación de Regularizaciones: Analizar cómo la regularización L1 y L2 afectan el desempeño del modelo, la selección de características y la capacidad de generalización.

Visualización de Resultados: Crear tablas que muestren las métricas de desempeño de cada modelo, facilitando la comparación y selección del mejor enfoque.

```python
# --- 1) Preparar los datos escalados originales (grado=1 lineal)
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_val_scaled   = scaler_X.transform(X_val)

# --- 2) Entrenar cada modelo con sus hiperparámetros óptimos
#    (aquí asumo que tienes best_deg, results['Lasso']['degree'], results['Lasso']['alpha'], etc.)

models = {}

# 2.1 Lineal puro
lr = LinearRegression().fit(X_train_scaled, y_train)
models['Linear'] = (X_train_scaled, X_val_scaled, lr)

# 2.2 Polinómico (grado óptimo best_deg)
poly = PolynomialFeatures(degree=best_deg, include_bias=False)
Xtr_poly = poly.fit_transform(X_train_scaled)
Xvl_poly = poly.transform(X_val_scaled)
scaler_pf = StandardScaler()
Xtr_pf_s = scaler_pf.fit_transform(Xtr_poly)
Xvl_pf_s = scaler_pf.transform(Xvl_poly)
lr_poly = LinearRegression().fit(Xtr_pf_s, y_train)
models['Polynomial'] = (Xtr_pf_s, Xvl_pf_s, lr_poly)

# 2.3 Lasso (grado y alpha óptimos)
deg_l, alpha_l = results['Lasso']['degree'], results['Lasso']['alpha']
poly_l = PolynomialFeatures(degree=deg_l, include_bias=False)
Xtr_l = poly_l.fit_transform(X_train_scaled)
Xvl_l = poly_l.transform(X_val_scaled)
scaler_l = StandardScaler()
Xtr_l_s = scaler_l.fit_transform(Xtr_l)
Xvl_l_s = scaler_l.transform(Xvl_l)
lasso = Lasso(alpha=alpha_l, max_iter=10000).fit(Xtr_l_s, y_train)
models['Lasso'] = (Xtr_l_s, Xvl_l_s, lasso)

# 2.4 Ridge (grado y alpha óptimos)
deg_r, alpha_r = results['Ridge']['degree'], results['Ridge']['alpha']
poly_r = PolynomialFeatures(degree=deg_r, include_bias=False)
Xtr_r = poly_r.fit_transform(X_train_scaled)
Xvl_r = poly_r.transform(X_val_scaled)
scaler_r = StandardScaler()
Xtr_r_s = scaler_r.fit_transform(Xtr_r)
Xvl_r_s = scaler_r.transform(Xvl_r)
ridge = Ridge(alpha=alpha_r, max_iter=10000).fit(Xtr_r_s, y_train)
models['Ridge'] = (Xtr_r_s, Xvl_r_s, ridge)

# --- 3) Función auxiliar para métricas
def compute_metrics(y_true, y_pred):
    mse  = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2   = r2_score(y_true, y_pred)
    # R = sqrt(R2) solo si R2 >= 0
    r    = np.sqrt(r2) if r2 >= 0 else np.nan
    return mse, rmse, r, r2

# --- 4) Calcular métricas y crear DataFrame
rows = []
for name, (Xtr, Xvl, mdl) in models.items():
    for split, (X, y) in [('train', (Xtr, y_train)), ('val', (Xvl, y_val))]:
        y_pred = mdl.predict(X)
        mse, rmse, r, r2 = compute_metrics(y, y_pred)
        rows.append({
            'Model' : name,
            'Split' : split,
            'MSE'    : mse,
            'RMSE'  : rmse,
            'R'     : r,
            'R2'    : r2
        })

df_metrics = pd.DataFrame(rows)
print(df_metrics)
```

### 5. Interpretación de Resultados y Conclusiones:

Identificación de Variables Clave: Determinar cuáles son las variables predictoras más importantes según los modelos entrenados, especialmente observando los coeficientes en modelos Lasso.

Impacto de la Regularización: Evaluar cómo la regularización contribuye a mejorar el modelo y prevenir el sobreajuste.

Recomendaciones Estratégicas: Proporcionar insights basados en los resultados obtenidos que puedan ayudar a la cadena de tiendas a optimizar sus estrategias de marketing, operaciones y gestión de recursos para maximizar las ventas.

```python
# 0) Variables seleccionadas
features = [
    'Num_Empleados',
    'Tamaño_Tienda',
    'Gasto_Publicidad_Online',
]
X_train_sub = X_train[features]
X_val_sub   = X_val[features]

# 1) Escalado de X
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train_sub)
X_val_scaled   = scaler_X.transform(X_val_sub)

# 2) Línea base: regresión lineal
lr_base = LinearRegression().fit(X_train_scaled, y_train)

# 3) Buscar mejor grado polinómico (2–5) para regresión lineal
degrees = [2, 3, 4, 5]
best_deg_poly, best_mse_poly = None, np.inf

for deg in degrees:
    poly = PolynomialFeatures(degree=deg, include_bias=False)
    Xtr_p = poly.fit_transform(X_train_scaled)
    Xvl_p = poly.transform(X_val_scaled)
    lr = LinearRegression().fit(Xtr_p, y_train)
    mse = mean_squared_error(y_val, lr.predict(Xvl_p))
    if mse < best_mse_poly:
        best_mse_poly, best_deg_poly = mse, deg

# 4) Optimizar Lasso y Ridge (grado 2–5, alphas dados)
alphas = [1e-3, 1e-2, 1e-1, 1, 10]
results = {}

for name, Model in [("Lasso", Lasso), ("Ridge", Ridge)]:
    best = {"degree": None, "alpha": None, "mse": np.inf}
    for deg in degrees:
        poly = PolynomialFeatures(degree=deg, include_bias=False)
        Xtr_p = poly.fit_transform(X_train_scaled)
        Xvl_p = poly.transform(X_val_scaled)
        # escalar espacio polinómico
        scaler_pf = StandardScaler()
        Xtr_ps = scaler_pf.fit_transform(Xtr_p)
        Xvl_ps = scaler_pf.transform(Xvl_p)
        for alpha in alphas:
            mdl = Model(alpha=alpha, max_iter=10_000).fit(Xtr_ps, y_train)
            mse = mean_squared_error(y_val, mdl.predict(Xvl_ps))
            if mse < best["mse"]:
                best.update({"degree": deg, "alpha": alpha, "mse": mse})
    results[name] = best

# 5) Preparar modelos finales con hiperparámetros óptimos
models = {}

# 5.1 Lineal puro
models['Linear'] = (X_train_scaled, X_val_scaled, lr_base)

# 5.2 Polinómico (grado optimal)
poly_best = PolynomialFeatures(degree=best_deg_poly, include_bias=False)
Xtr_poly   = poly_best.fit_transform(X_train_scaled)
Xvl_poly   = poly_best.transform(X_val_scaled)
scaler_pf  = StandardScaler()
Xtr_ps     = scaler_pf.fit_transform(Xtr_poly)
Xvl_ps     = scaler_pf.transform(Xvl_poly)
lr_poly    = LinearRegression().fit(Xtr_ps, y_train)
models['Polynomial'] = (Xtr_ps, Xvl_ps, lr_poly)

# 5.3 Lasso (grado y alpha óptimos)
deg_l, alpha_l = results['Lasso']['degree'], results['Lasso']['alpha']
poly_l = PolynomialFeatures(degree=deg_l, include_bias=False)
Xtr_l = poly_l.fit_transform(X_train_scaled)
Xvl_l = poly_l.transform(X_val_scaled)
scaler_l = StandardScaler()
Xtr_ls  = scaler_l.fit_transform(Xtr_l)
Xvl_ls  = scaler_l.transform(Xvl_l)
lasso = Lasso(alpha=alpha_l, max_iter=10_000).fit(Xtr_ls, y_train)
models['Lasso'] = (Xtr_ls, Xvl_ls, lasso)

# 5.4 Ridge (grado y alpha óptimos)
deg_r, alpha_r = results['Ridge']['degree'], results['Ridge']['alpha']
poly_r = PolynomialFeatures(degree=deg_r, include_bias=False)
Xtr_r = poly_r.fit_transform(X_train_scaled)
Xvl_r = poly_r.transform(X_val_scaled)
scaler_r = StandardScaler()
Xtr_rs  = scaler_r.fit_transform(Xtr_r)
Xvl_rs  = scaler_r.transform(Xvl_r)
ridge = Ridge(alpha=alpha_r, max_iter=10_000).fit(Xtr_rs, y_train)
models['Ridge'] = (Xtr_rs, Xvl_rs, ridge)

# 6) Función para métricas
def compute_metrics(y_true, y_pred):
    mse  = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2   = r2_score(y_true, y_pred)
    r    = np.sqrt(r2) if r2 >= 0 else np.nan
    return mse, rmse, r, r2

# 7) Calcular métricas train/val y montar DataFrame
rows = []
for name, (Xtr, Xvl, mdl) in models.items():
    for split, (X, y) in [('train', (Xtr, y_train)), ('val', (Xvl, y_val))]:
        y_pred      = mdl.predict(X)
        mse, rmse, r, r2 = compute_metrics(y, y_pred)
        rows.append({
            'Model' : name,
            'Split' : split,
            'MSE'    : mse,
            'RMSE'  : rmse,
            'R'     : r,
            'R2'    : r2
        })

df_metrics = pd.DataFrame(rows)
print(df_metrics)
```
